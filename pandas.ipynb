```python
#Data Cleaning & Preprocessing
#Sample Data (Dirty Data)
import pandas as pd
import numpy as np

data = {
    'Name': ['Ram', 'Shyam', 'Hari', 'Ram', None],
    'Age': [21, np.nan, 23, 21, 22],
    'Marks': [85, 90, np.nan, 85, 88],
    'City': ['Hyderabad', 'Delhi', 'Mumbai', 'Hyderabad', None]
}

df = pd.DataFrame(data)
print(df)
```

```python
#1Ô∏è‚É£ Detect Missing Values
#Check missing values
df.isnull()
```

```python
#Count missing values
df.isnull().sum()
```

```python
#Check non-missing values
df.notnull()
```

```python
#2Ô∏è‚É£ Handling Missing Values
#Drop rows with missing values
df.dropna()
```

```python
#Drop columns with missing values
df.dropna(axis=1)
```

```python
#Fill missing values with a constant
df.fillna(0)
```

```python
#Fill with mean / median / mode
df['Age'].fillna(df['Age'].mean())
df['Marks'].fillna(df['Marks'].median())
df['City'].fillna(df['City'].mode()[0])
```

```python
#3Ô∏è‚É£ Remove Duplicate Data
#Check duplicates
df.duplicated()
```

```python
#Remove duplicates
df.drop_duplicates()
```

```python
#4Ô∏è‚É£ Rename Columns
df.rename(columns={'Marks': 'Score', 'City': 'Location'})
```

```python
#5Ô∏è‚É£ Change Data Types
#Check data types
df.dtypes
```

```python
#6Ô∏è‚É£ String Cleaning (Text Data)
df['Name'] = df['Name'].str.lower()
df['City'] = df['City'].str.strip()
```

```python
df['City'] = df['City'].str.replace('Hyderabad', 'HYD')
```

```python
#7Ô∏è‚É£ Handling Incorrect Values
#Replace wrong values
df['Marks'] = df['Marks'].replace(-1, np.nan)
```

```python
#8Ô∏è‚É£ Outlier Handling
#Using condition
df[df['Marks'] < 100]
```

```python
#Remove outliers
df = df[df['Marks'] >= 40]
```

```python
#9Ô∏è‚É£ Data Normalization / Scaling
#Min-Max Normalization
df['Marks_norm'] = (df['Marks'] - df['Marks'].min()) / (df['Marks'].max() - df['Marks'].min())
```

```python
#üîü Encoding Categorical Data
#Label Encoding
df['City_code'] = df['City'].astype('category').cat.codes
```

```python
#One-Hot Encoding
pd.get_dummies(df, columns=['City'])
```

```python
#1Ô∏è‚É£1Ô∏è‚É£ Filtering Clean Data
df_clean = df[(df['Age'] > 20) & (df['Marks'] > 80)]
```

```python
import pandas as pd

# DataFrame 1
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Ram', 'Shyam', 'Hari']
})

print("df1:")
print(df1)

# DataFrame 2
df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Marks': [85, 90, 88]
})

print("\ndf2:")
print(df2)
```

```python
#üîó CONCAT Example
#Vertical Concatenation
result_concat = pd.concat([df1, df2], axis=0)
print(result_concat)
```

```python
#üîó MERGE Example (Most Important)
result_merge = pd.merge(df1, df2, on='ID')
print(result_merge)
```

```python
#üîó MERGE with Join Types
pd.merge(df1, df2, on='ID', how='left')
pd.merge(df1, df2, on='ID', how='right')
pd.merge(df1, df2, on='ID', how='outer')
```

```python
#üîó JOIN Example (Index-based)
df1_indexed = df1.set_index('ID')
df2_indexed = df2.set_index('ID')

result_join = df1_indexed.join(df2_indexed)
print(result_join)
```
